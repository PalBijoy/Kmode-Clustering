```python
#Mining of Road Accident using Kmode Clustering
```

```python
import pandas as pd
import numpy as np
import math
from kmodes.kmodes import KModes
```

```python
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
plt.scatter(df['Place'],df['Number of Accident'])
```

```python
km = KModes(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Place','Number of Accident']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Place,df1['Number of Accident'],color='red')

plt.xlabel('Place')
plt.ylabel('Number of Accident')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Number of Accident']])
df
```

```python
km = KModes(n_clusters=3)
y_predicted = km.fit_predict(df[['Place','Number of Accident']])
y_predicted
```

```python
df['cluster'] = y_predicted
#df.drop('cluster',axis='columns',inplace=True)
df
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1.Place,df1['Number of Accident'],color='green')

plt.xlabel('Place')
plt.ylabel('Number of Accident')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KModes(n_clusters=k)
    km.fit(df[['Place','Number of Accident']])
    sse.append(km)
```
